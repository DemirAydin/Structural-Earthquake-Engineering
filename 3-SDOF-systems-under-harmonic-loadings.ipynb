## 1.0) Free Vibration of SDOF Systems

### 1.1) Undamped System

<img src="./Figures/Free_vibration_undamped.png" align="left"/>

```python
import numpy as np
import matplotlib.pyplot as plt

# Input
k = 100      # (kN/m); Stiffness
m = 50       # (ton); Mass
u0 = 1.0    # (m); Initial displacement
v0 = 1.0    # (m/s); Initial velocity

# Calculations * DO NOT CHANGE!
wn = np.sqrt(k/m)      # (rad/s); Natural angular frequency
Tn = 2 * np.pi / wn    # (s); Natural period
fn = 1 / Tn            # (1/s) or (Hz); Natural frequency
t = np.arange(0, 10, Tn/100)

A = u0
B = v0 / wn
u = A * np.cos(wn * t) + B * np.sin(wn * t)

print("fn =", fn, "Hz")
print("wn =", wn, "rad/s")

# Plot Graph
fig, ax = plt.subplots()
ax.plot(t, u, color="r")
ax.set(xlabel='Time (s)', ylabel='u (m)', title='t - u Curve')
ax.grid(True)

plt.show()
```

### 1.2) Damped System

<img src="./Figures/Free_vibration_damped.png" align="left"/>

```python
import numpy as np
import matplotlib.pyplot as plt

# Input
Tn = 0.2                    # System natural period (s)
ksi = 0.050                  # Damping ratio / for undamped system: ksi = 0
u0 = 1.0                    # Initial displacement (m)
v0 = 5.0                    # Initial velocity (m/s)
t = np.arange(0, 5, Tn/100) # Time axis

# Calculations * DO NOT CHANGE!
wn = 2 * np.pi / Tn               # Natural angular frequency (rad/s)
wd = wn * np.sqrt(1 - ksi**2)     # Damped angular frequency (rad/s)

print("Wn =", wn, "rad/s", "  Wd =", wd, "rad/s")

A = u0
B = (v0 + ksi * wn * u0) / wd
u = np.exp(-ksi * wn * t) * (A * np.cos(wd * t) + B * np.sin(wd * t))

# Plot Graph
fig, ax = plt.subplots()
ax.plot(t, u, color="red", label="Tn = {} s, ξ = {}%".format(Tn, ksi * 100))
ax.set(xlabel='Time (s)', ylabel='u (m)', title='t - u curve')
ax.grid(True)
ax.legend()

plt.show()
```

## 2.0) Forced Vibration (Harmonic Motion) of SDOF Systems

### 2.1) Undamped System:

<img src="./Figures/Harmonic_vibration_undamped.png" align="left"/>

```python
import numpy as np
import matplotlib.pyplot as plt

# Input
T=0.9        # (s); Excitation Period
Tn=0.3       # (s); System Natural Period
p0=1000.0     # (kN); Excitation Force Amplitude
u0=0.0       # (m); Initial Displacement
v0=2         # (m/s); Initial Velocity 
m=5.0        # (ton); Assumption for Mass
t=np.arange(0,10,Tn/100)   # (s); Time axis

# Calculations * DO NOT CHANGE!
w=2*np.pi/T                 # (rad/s); Excitation Angular Freq.
wn=2*np.pi/Tn               # (rad/s); Natural angular Frequency
b=w/wn                      # Frequency Ratio
k=m*wn**2                   # (kN/m); System Stiffness
ust=p0/k                    # (m); Static displacement

# Conplementary solution, transient vibration
B = u0
A = (v0/wn)-(ust*b/(1-b**2))
uc = A*np.sin(wn*t)+B*np.cos(wn*t)

# Particular solution, Steady-state vibration
D = ust*(1/(1-b**2))
E = 0
up = D*np.sin(w*t)+E*np.cos(w*t)

# General Solution
ut = uc+up

# Plot Graphs
fig, axes = plt.subplots(nrows=3, ncols=1)

axes[0].set_title("t - uc grafiği")
axes[0].plot(t, uc, color='g')
axes[0].set_xlabel("Time (s)")
axes[0].set_ylabel("uc (m)")
axes[0].grid()

axes[1].set_title("t - up grafiği")
axes[1].plot(t, up, color='b')
axes[1].set_xlabel("Time (s)")
axes[1].set_ylabel("up (m)")
axes[1].grid()

axes[2].set_title("t - ut grafiği")
axes[2].plot(t, ut, color='r')
axes[2].set_xlabel("Time (s)")
axes[2].set_ylabel("ut (m)")
axes[2].grid()

fig.tight_layout()    # Grafiklerin iç içe geçmemesi için
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Input
T = 0.9        # (s); Excitation period
Tn = 0.3       # (s); System natural period
p0 = 1000.0    # (kN); Excitation force amplitude
u0 = 0.0       # (m); Initial displacement
v0 = 2.0       # (m/s); Initial velocity
m = 5.0        # (ton); Assumed mass
t = np.arange(0, 10, Tn/100)   # (s); Time axis

# Calculations * DO NOT CHANGE!
w = 2 * np.pi / T              # (rad/s); Excitation angular frequency
wn = 2 * np.pi / Tn            # (rad/s); Natural angular frequency
b = w / wn                     # Frequency ratio
k = m * wn**2                  # (kN/m); System stiffness
ust = p0 / k                   # (m); Static displacement

# Complementary solution (transient vibration)
B = u0
A = (v0 / wn) - (ust * b / (1 - b**2))
uc = A * np.sin(wn * t) + B * np.cos(wn * t)

# Particular solution (steady-state vibration)
D = ust * (1 / (1 - b**2))
E = 0
up = D * np.sin(w * t) + E * np.cos(w * t)

# General solution
ut = uc + up

# Plot Graphs
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(6, 7))

axes[0].set_title("t - uc curve")
axes[0].plot(t, uc, color='g')
axes[0].set_xlabel("Time (s)")
axes[0].set_ylabel("uc (m)")
axes[0].grid(True)

axes[1].set_title("t - up curve")
axes[1].plot(t, up, color='b')
axes[1].set_xlabel("Time (s)")
axes[1].set_ylabel("up (m)")
axes[1].grid(True)

axes[2].set_title("t - ut curve")
axes[2].plot(t, ut, color='r')
axes[2].set_xlabel("Time (s)")
axes[2].set_ylabel("ut (m)")
axes[2].grid(True)

fig.tight_layout()   # Prevent overlapping plots
plt.show()
```

### 2.2) Damped System:

<img src="./Figures/Harmonic_vibration_damped.png" align="left"/>

```python
import numpy as np
import matplotlib.pyplot as plt
print("Plot damped SDOF system response to harmonic force")

# Input
T = 0.8        # (s); Excitation period
Tn = 0.5       # (s); System natural period
ksi = 0.05     # Damping ratio / for undamped system: ksi = 0
p0 = 100.0     # (kN); Excitation force amplitude
u0 = 0.1       # (m); Initial displacement
v0 = 0.5       # (m/s); Initial velocity
m = 50.0       # (ton); Assumed mass
t = np.arange(0, 10, Tn/100)   # (s); Time axis

# Calculations * DO NOT CHANGE!
w = 2 * np.pi / T                    # (rad/s); Excitation angular frequency
wn = 2 * np.pi / Tn                  # (rad/s); Natural angular frequency
wd = wn * np.sqrt(1 - ksi**2)        # (rad/s); Damped angular frequency
b = w / wn                           # Frequency ratio
k = m * wn**2                        # (kN/m); System stiffness
ust = p0 / k                         # (m); Static displacement
bb = (1 - b**2)**2 + (2 * ksi * b)**2   # A fictitious coefficient to simplify the following equations

# Complementary solution; transient vibration
A = (v0 / wd) + (ksi * wn * u0 / wd) + (2 * ksi**2 * wn * b * ust / (bb * wd)) - ((1 - b**2) * w * ust / (bb * wd))
B = u0 + 2 * ksi * b * ust / bb
uc = np.exp(-ksi * wn * t) * (A * np.sin(wd * t) + B * np.cos(wd * t))

# Particular solution; steady-state vibration
D = ust * (1 - b**2) / bb
E = -2 * ksi * b * ust / bb
up = D * np.sin(w * t) + E * np.cos(w * t)

# General solution
ut = uc + up

# Plot Graphs
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 8))

axes[0].set_title("t - uc curve")
axes[0].plot(t, uc, color='g')
axes[0].set_xlabel("Time (s)")
axes[0].set_ylabel("uc (m)")
axes[0].grid(True)

axes[1].set_title("t - up curve")
axes[1].plot(t, up, color='b')
axes[1].set_xlabel("Time (s)")
axes[1].set_ylabel("up (m)")
axes[1].grid(True)

axes[2].set_title("t - ut curve")
axes[2].plot(t, ut, color='r')
axes[2].set_xlabel("Time (s)")
axes[2].set_ylabel("ut (m)")
axes[2].grid(True)

fig.tight_layout()
plt.show()
```
